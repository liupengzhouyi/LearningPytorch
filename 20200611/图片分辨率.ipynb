```
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = np.array(Image.open('/content/drive/My Drive/壁纸/壁纸——钱.png'))

rows,cols,dims=img.shape
for i in range(10000):
    x=np.random.randint(0,rows)
    y=np.random.randint(0,cols)
    img[x,y,:]=255

plt.figure("love")
plt.imshow(img)
plt.axis('off')
plt.show()
```

```
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = np.array(Image.open('/content/drive/My Drive/壁纸/壁纸——钱.png').convert('L'))

rows,cols=img.shape
for i in range(rows):
    for j in range(cols):
        if (img[i,j]<=128):
            img[i,j]=0
        else:
            img[i,j]=1

plt.figure("love")
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = Image.open('/content/drive/My Drive/壁纸/壁纸——钱.png')

box = img.copy()
box = (2333, 800, 6000, 3800)
region = img.crop(box)

region.save('/content/drive/My Drive/壁纸/壁纸——钱1.png', "JPEG")

region = region.resize((28, 28),Image.ANTIALIAS)

region.save('/content/drive/My Drive/壁纸/壁纸——钱2.png', "JPEG")

plt.figure("love")
plt.imshow(region, cmap='gray')
plt.axis('off')
plt.show()
```
